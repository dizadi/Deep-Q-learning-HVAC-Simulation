```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("GatesData8-29.csv")
```

```python
def add_runtime(data):
    
    request_status = data['Chilled Water Request']
    run_time = np.zeros(request_status.shape[0])
    t=0

    for i in range(request_status.shape[0]):
        status = request_status[i]
        if status == 0:
            t=0
        elif status==1:
            t+=status
        run_time[i] = t

    data['Run Time'] = run_time
    
    return data



flow_parameters = ['Entry', 'CHWRT', 'CHWST', 'Chilled Water Request', 'OA Humidity',
       'OA Temperature', '3rd Floor AHU-1|SA Airflow',
       '3rd Floor AHU-2|SA Airflow', '3rd Floor AHU-3|SA Airflow',
       '3rd Floor AHU-7|SA Airflow', '3rd Floor AHU-8|SA Airflow',
       'Level 2 AHU-4|SA Airflow', 'Level 2 AHU-5|SA Airflow',
       'Roof AHU-10|SA Airflow', 'Roof AHU-11|SA Airflow',
       'Roof AHU-12|SA Airflow', 'Roof AHU-9|SA Airflow',
       '3rd Floor AHU-1|RA Humidity', '3rd Floor AHU-2|RA Humidity',
       '3rd Floor AHU-3|RA Humidity', '3rd Floor AHU-7|RA Humidity',
       '3rd Floor AHU-8|RA Humidity', 'Level 2 AHU-4|RA Humidity',
       'Level 2 AHU-5|RA Humidity', 'Roof AHU-10|RA Humidity',
       'Roof AHU-11|RA Humidity', 'Roof AHU-12|RA Humidity',
       'Roof AHU-9|RA Humidity', 'Run Time', 'CHW Flowrate']

data[flow_parameters].iloc[0].values
```

```python
data = data.rename({'Entry':'Time'}, axis=1)
flowrate = data['CHW Flowrate']
return_temp = data['CHWRT']
action = data['Chilled Water Request']
power = data['Apparent Power']
data = data.drop(columns = ['Unnamed: 0','CHW Flowrate', 'CHWRT', 'Chilled Water Request', 'Apparent Power'])
data['Flow Rate'] = np.zeros(len(data.index))
data['Return Temperature'] = np.zeros(len(data.index))
data['Action'] = np.zeros(len(data.index))
data['Power'] = np.nan
data['Power Actual'] = power
```
